# Práctica: Interprete JS con síntesis de voz

En esta práctica, vamos a crear una página web simple utilizando HTML, CSS y JavaScript que permite al usuario ingresar texto y reproducirlo en voz alta mediante la síntesis de voz.

## Instrucciones

1. **Configura tu entorno de trabajo:**

   - Crea una nueva carpeta para tu proyecto.
   - Dentro de esta carpeta, crea los siguientes archivos:
     - `index.html`
     - `style.css`
     - `script.js`
